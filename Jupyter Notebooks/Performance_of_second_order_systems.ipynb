# Performance of Second-order Systems

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import control
```

```python
# Define python function for step response of second-order system
def step_response(A,wn,zeta,t):
    '''
    A - step input magnitude
    wn - natural frequency
    zeta - damping ratio
    t - time array'''

    wd = wn*np.sqrt(1-zeta**2)
    phi = np.arccos(zeta)

    y = A - A*(1/np.sqrt(1-zeta**2))*np.exp(-zeta*wn*t)*np.sin(wd*t+phi)

    return y

# Define python function for impulse response of second-order system
def impulse_response(A,wn,zeta,t):
    '''
    A - impulse input magnitude
    wn - natural frequency
    zeta - damping ratio
    t - time array'''

    wd = wn*np.sqrt(1-zeta**2)

    y = A*(wn**2/np.sqrt(1-zeta**2))*np.exp(-zeta*wn*t)*np.sin(wd*t)

    return y
```

## Step Input

```python
A = 1 # unit step input
wn = 2
zeta1 = 0
zeta2=0.1
zeta3=0.3
zeta4=0.7

t = np.arange(0,10,0.05)

y1_step = step_response(A,wn,zeta1,t)
y2_step = step_response(A,wn,zeta2,t)
y3_step = step_response(A,wn,zeta3,t)
y4_step = step_response(A,wn,zeta4,t)
```

```python
#############################################################
fig = plt.figure(figsize=(6,4))
# fig = plt.figure(figsize=(12,8))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17, left=0.17, top=0.96, right=0.96)

# Change the axis units font
plt.setp(ax.get_ymajorticklabels(),fontsize=18)
plt.setp(ax.get_xmajorticklabels(),fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':', color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
plt.xlabel(r'Normalized Time $(\omega_n t)$', fontsize=22, weight='bold', labelpad=5)
plt.ylabel(r'$y(t)$', fontsize=22, weight='bold', labelpad=10)

plt.plot(wn*t, y1_step, linewidth=2, linestyle='-', label=r'$\zeta=0$')
plt.plot(wn*t, y2_step, linewidth=2, linestyle='--', label=r'$\zeta=0.1$')
plt.plot(wn*t, y3_step, linewidth=2, linestyle='-.', label=r'$\zeta=0.3$')
plt.plot(wn*t, y4_step, linewidth=2, linestyle=':', label=r'$\zeta=0.7$')
# plt.plot(time, np.exp(-zeta*wn*time), linewidth=1, linestyle='--', color='#377eb8')
# plt.plot(time, -np.exp(-zeta*wn*time), linewidth=1, linestyle='--', color='#377eb8')
# uncomment below and set limits if needed
# plt.xlim(0,5)
plt.ylim(None,3)

# Create the legend, then fix the fontsize
leg = plt.legend(loc='upper right', ncol = 2, fancybox=True, )
ltext  = leg.get_texts()
plt.setp(ltext,fontsize=16)

# Adjust the page layout filling the page using the new tight_layout command
plt.tight_layout(pad=0.5)

# save the figure as a high-res pdf in the current folder
# plt.savefig('plot_filename.pdf')

# plt.show()
```

## Impulse Input

```python
A = 1 # unit impulse input
wn = 2
zeta1 = 0
zeta2=0.1
zeta3=0.3
zeta4=0.7

t = np.arange(0,10,0.05)

y1_impulse = impulse_response(A,wn,zeta1,t)
y2_impulse = impulse_response(A,wn,zeta2,t)
y3_impulse = impulse_response(A,wn,zeta3,t)
y4_impulse = impulse_response(A,wn,zeta4,t)
```

```python
#############################################################
fig = plt.figure(figsize=(6,4))
# fig = plt.figure(figsize=(12,8))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17, left=0.17, top=0.96, right=0.96)

# Change the axis units font
plt.setp(ax.get_ymajorticklabels(),fontsize=18)
plt.setp(ax.get_xmajorticklabels(),fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':', color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
plt.xlabel(r'Normalized Time $(\omega_n t)$', fontsize=22, weight='bold', labelpad=5)
plt.ylabel(r'$y(t)/\omega_n^2$', fontsize=22, weight='bold', labelpad=10)

plt.plot(wn*t, y1_impulse/wn**2, linewidth=2, linestyle='-', label=r'$\zeta=0$')
plt.plot(wn*t, y2_impulse/wn**2, linewidth=2, linestyle='--', label=r'$\zeta=0.1$')
plt.plot(wn*t, y3_impulse/wn**2, linewidth=2, linestyle='-.', label=r'$\zeta=0.3$')
plt.plot(wn*t, y4_impulse/wn**2, linewidth=2, linestyle=':', label=r'$\zeta=0.7$')
# plt.plot(time, np.exp(-zeta*wn*time), linewidth=1, linestyle='--', color='#377eb8')
# plt.plot(time, -np.exp(-zeta*wn*time), linewidth=1, linestyle='--', color='#377eb8')
# uncomment below and set limits if needed
# plt.xlim(0,5)
plt.ylim(None,2)

# Create the legend, then fix the fontsize
leg = plt.legend(loc='upper right', ncol = 2, fancybox=True, )
ltext  = leg.get_texts()
plt.setp(ltext,fontsize=16)

# Adjust the page layout filling the page using the new tight_layout command
plt.tight_layout(pad=0.5)

# save the figure as a high-res pdf in the current folder
# plt.savefig('plot_filename.pdf')

# plt.show()
```
