# P7.5
## (a)

The closed-loop TF for the automatic control loop ignoring the disturbance (for now) is:

$$T(s)=\frac{G(s)}{1+G(s)\frac{K_2(s+1)}{s+9}}$$

where:

$$G(s)=\frac{25(s+0.03)}{(s+0.4)(s^2-0.36s+0.16)}$$

The characteristic equation that we will use for the root locus is:

$$1+K_s\frac{25(s+0.03)}{(s+0.4)(s^2-0.36s+0.16)}\frac{(s+1)}{s+9}=0$$

```python
sys_auto = (25*(s+0.03)*(s+1))/((s+0.4)*(s**2-0.36*s+0.16)*(s+9))

_ = control.root_locus(sys_auto)
```
